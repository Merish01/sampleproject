```python
import numpy as np
marks = np.array([45, 78, 56, 89, 33, 67, 90, 52])

print("Student Marks:", marks)

average = np.mean(marks)
maximum = np.max(marks)
minimum = np.min(marks)

print("\nAverage Marks:", average)
print("Highest Marks:", maximum)
print("Lowest Marks:", minimum)

pass_marks = marks[marks >= 50]
fail_marks = marks[marks < 50]

print("\nPassed Students Marks:", pass_marks)
print("Failed Students Marks:", fail_marks)

print("\nNumber of Passed Students:", len(pass_marks))
print("Number of Failed Students:", len(fail_marks))
```
